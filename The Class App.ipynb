```python
class Student:
    def __init__(self, name):
        self.name = name
        self.grades = {"Midterm": None, "Homework": None, "Final": None}

    def add_grade(self, grade, input_mark):
        self.grades[grade] = input_mark

    def average(self):
        grades = [n for n in self.grades.values() if n is not None]
        return sum(grades) / len(grades) if grades else None

    def __repr__(self):
        return f"{self.name}: {self.grades}"
    
students = []

while True:
    print("The Class App:")
    print("1 : Your Students ")
    print("2 : Add Student")
    print("3 : Input Mark")
    print("4 : Class Details")
    print("else : Exit")

    option = input("What is your choice?\n")
    
    if option == "1":
        print("Your Class:")
        for student in students:
            print(student)
        
    elif option == "2":
        name = input("What is student name? ")
        student = Student(name)
        students.append(student)
        print("The student has been saved!")
        
    elif option == "3":
        if not students:
            print("There is no Student in the class!")
            continue

        name = input("What is student name who add/change the marks? ")

        try:
            student = next(i for i in students if i.name == name)
        except StopIteration:
            print("The student has not been founded!")
            continue

        for grade in ["Midterm", "Homework", "Final"]:
            input_mark = input(f"{grade} Grade (you can pass by press space): ")
            
            if not input_mark:
                continue

            try:
                input_mark = float(input_mark)
            except ValueError:
                print("Please enter valid grade not a string")
                continue

            student.add_grade(grade, input_mark)

        print("The Grades has been saved successfully!")
        
    elif option == "4":
        print("Your Class:")
        for student in students:
            print(student)
        student_num = len(students)
        Average = sum(student.average() for student in students if student.average() is not None) / student_num if student_num > 0 else 0

        print(f"\nTotal Student number: {student_num}")
        print(f"Class Average: {Average}")    
    else:
        print("Exiting...")
        print("The Program has ended!")
        break
```
